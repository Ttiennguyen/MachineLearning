```python
#Cubic polynomial
yr1 = [1971]
yr2 = [1972]
yr3 = [1973]
yr4 = [1974]
yr5 = [1975]
yr6 = [1976]
yr7 = [1977]
yr8 = [1978]
yr9 = [1979]
yr10 = [1980]

import numpy as np
X = np.array( [yr1, yr2,yr3,yr4,yr5,yr6,yr7,yr8, yr9, yr10])
y = np.array([45.13, 51.71,60.17,64.83,65.24,65.17,67.65,79.80, 96.13, 115.19])
print (X.shape, y.shape)

#testing set
yr11 = [1981]
yr12 = [1982]
X_test = np.array([yr11, yr12])
y_test = np.array([130, 200])
print (X_test.shape, y_test.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, c='r', label='train data')
plt.scatter(X_test, y_test, c='g', label='test data')
plt.xlabel('Year')
plt.ylabel('Cost')
plt.legend()
plt.show()
```

```python
# training
from sklearn import preprocessing
poly = preprocessing. PolynomialFeatures (degree=3, include_bias=False)
Xp = poly.fit_transform (X)
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(Xp, y)
# evaluation
from sklearn import metrics
Xp_test = poly.fit_transform (X_test)
y_test_hat = model.predict (Xp_test)
print (y_test_hat)
print (metrics.mean_squared_error (y_test, y_test_hat))
print (metrics.r2_score (y_test, y_test_hat))
```
