```python
# Dataset
p1 = [1]
p2 = [2]
p3 = [3]
p4 = [4]
p5 = [5]
p6 = [6]
p7 = [7]
p8 = [8]
p9 = [9]
p10 = [10]
p11 = [11]
p12 = [12]
p13 = [13]
p14 = [14]
p15 = [15]
p16 = [16]
p17 = [17]
p18 = [18]
p19 = [19]
p20 = [20]
p21 = [21]
p22 = [22]
p23 = [23]
p24 = [24]
p25 = [25]
p26 = [26]
p27 = [27]
p28 = [28]
p29 = [29]
p30 = [30]

import numpy as np

X = np.array([p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, p11, p12, p13, p14, p15, p16, p17, p18, p19, p20, p21, p22, p23, p24, p25, p26, p27, p28, p29, p30])
y = np.array([3, 4, 5, 7, 10, 8, 9, 10, 10, 23, 27, 44, 50, 63, 67, 60, 62, 70, 75, 88, 81, 87, 95, 100, 108, 135, 151, 160, 169, 179])

print(X.shape, y.shape)

p31 =[31]
p32 =[32]
X_test = np.array([p31, p32])
y_test = np.array([180, 200])
print(X_test.shape, y_test.shape)
```

```python
# visualization of dataset
import matplotlib.pyplot as plt
plt.scatter(X,y,c='r',label='train data')
plt.scatter(X_test,y_test,c='g',label='test data')
plt.legend()
plt.show()
```

```python
# training
# step 1 
from sklearn import preprocessing
poly = preprocessing.PolynomialFeatures(degree=2, include_bias=False)
# step 2
Xp = poly.fit_transform(X)
# step 3
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(Xp, y)
# evalaution
from sklearn import metrics
Xp_test = poly.fit_transform(X_test)
y_test_hat = model.predict(Xp_test)
print(y_test_hat)
print(metrics.mean_squared_error(y_test, y_test_hat))
print(metrics.r2_score(y_test, y_test_hat))
```

```python
#fitting y =ax^2 + bx + c
a = model.coef_[1]
b = model.coef_[0]
c = model.intercept_
#drawing line
xline = np.concatenate((X, X_test))
yline = a*xline**2 + b*xline + c
plt.plot(xline,yline,'k')
plt.scatter(X,y,c='r',label='train data')
plt.scatter(X_test,y_test,c='g',label='test data')
plt.scatter(X_test,y_test_hat,c='k',label='predicted data')
plt.legend()
plt.show()
```
