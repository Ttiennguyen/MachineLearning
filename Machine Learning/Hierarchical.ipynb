```python
# dataset
p1 = [4,21]
p2 = [5,19]
p3 = [10,24]
p4 = [4,17]
p5 = [3,16]
p6 = [11,25]
p7 = [14,24]
p8 = [6,22]
p9 = [10,21]
p10 = [12,21]
import numpy as np
X = np.array([p1,p2,p3,p4,p5,p6,p7,p8,p9,p10])
# visualization 
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1], c='k')
plt.show()
```

```python
# visualization
import scipy.cluster.hierarchy as shc
linkage = shc.linkage(X, method='ward', metric='euclidean')
shc.dendrogram(linkage)
plt.axhline(y=10, color='r', linestyle='dashed')
```

```python
# training
from sklearn import cluster
model = cluster.KMeans(n_clusters=2, n_init ='auto')
labels = model.fit_predict(X)
print(labels)
# evaluation
from sklearn import metrics
print(metrics.silhouette_score(X,labels))
```

```python
# visualization
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1], c=labels)
plt.show()
```
