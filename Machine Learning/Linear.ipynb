```python
p1 = [147]
p2 = [150]
p3 = [153]
p4 = [155]
p5 = [158]
p6 = [160]
p7 = [163]
p8 = [165]
p9 = [168]
p10 = [170]
p11 = [173]
p12 = [175]
p13 = [178]
p14 = [180]
p15 = [183]

import numpy as np

X = np.array([p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, p11, p12, p13, p14, p15])
y = np.array([49, 50, 51, 52, 54, 56, 58, 59, 60, 62, 63, 64, 66, 67, 68])
print(X.shape, y.shape)
p16 = [156]
p17 = [176]

X_test = np.array([p16, p17])
y_test = np.array([52, 66])

print(X_test.shape, y_test.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, c='r', label='train data')
plt.scatter(X_test, y_test, c='g', label='test data')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend()
plt.show()
```

```python
#training
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X, y)
#evaluation
from sklearn import metrics
y_test_hat = model.predict(X_test)
print(y_test_hat)
print(metrics.mean_squared_error(y_test, y_test_hat))
print(metrics.r2_score(y_test, y_test_hat))
```

```python
#fitting line
a = model.coef_[0]
b = model.intercept_
#draw fitting line
xline = np.concatenate((X, X_test))
yline = a * xline + b
plt.plot(xline, yline , 'k')
plt.scatter(X, y, c='r', label='train data')
plt.scatter(X_test, y_test, c='g', label='test data')
plt.scatter(X_test, y_test_hat, c='k', label='predict data')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend()
plt.show()
```
